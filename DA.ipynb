```python
# Created a data store file method to store what al insights i would skim from my DA here.
def txt(content, bullets=False,endline=True):
    
    txt_file=open('test.txt',"+a")
    if bullets==True:
        txt_file.write("-"+" "+content)
        txt_file.write("\n")
        if endline==True:
            end()
    elif type(bullets)==str:
        txt_file.write(bullets+" "+content)
        txt_file.write("\n")
        if endline==True:
            end()
    elif bullets==False or bullets==None:
        txt_file.write(content)
        txt_file.write("\n")
        if endline==True:
            end()
    
    txt_file.close()
def end():
    txt_file=open('test.txt',"+a")
    txt_file.write("_"*100+"\n\n")
    txt_file.close()
```
